# How are in-demand skills for Data Analysts in the U.S.?

### Methodology

1. Aggregate skill counts monthly
2. Re-analyze based on percentage of total jobs
3. Plot the monthly skill demand

```python
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='United States')].copy()
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills',aggfunc='size', fill_value=0)

df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_US_pivot = df_DA_US_pivot.drop('Total')
```

```python
df_DA_US_pivot
```

```python
DA_totals = df_DA_US.groupby('job_posted_month_no').size()

df_DA_US_perc = df_DA_US_pivot.div(DA_totals/100, axis=0)
```

```python
df_DA_US_perc = df_DA_US_perc.reset_index()
df_DA_US_perc['job_posted_month'] = df_DA_US_perc['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_perc = df_DA_US_perc.set_index('job_posted_month')
df_DA_US_perc = df_DA_US_perc.drop(columns='job_posted_month_no')

df_DA_US_perc
```

```python
df_plot = df_DA_US_perc.iloc[:, :5]

sns.set_theme(style='ticks')
sns.lineplot(data= df_plot, dashes=False, palette='tab10', legend=False)
sns.despine()

plt.title('Trending Top 5 Skills for Data Analyst in the US')
plt.ylabel('Likeilhood in Job Posting')
plt.xlabel('2023')

from matplotlib.ticker import PercentFormatter
from adjustText import adjust_text
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))
texts = []
for i in range(5):
    txt = plt.text(
        11.7,                    
        df_plot.iloc[-1, i],            
        df_plot.columns[i],             
        fontsize=9,
        va='center'
    )
    texts.append(txt)

adjust_text(texts, arrowprops=dict(arrowstyle='->', color='gray', lw=1))

plt.show()
```
